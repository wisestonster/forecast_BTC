```python
# 프로핏으로 로또 매직넘버 예측하기
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from fbprophet import Prophet

# prophet을 사용하기 위해서는 다음과 같이 피처의 이름을 변경해야 합니다 : 'ds', 'y'
csvFilename = input('lotto_win_info_숫자를 입력해 주세요: ')
print('로또 매직넘버를 예측합니다...')
full_df = pd.read_csv('lotto_win_info_{}.csv'.format(csvFilename))
df = full_df[['date', 'magicNo']]
df.columns = ['ds', 'y']
prophet = Prophet(seasonality_mode='multiplicative',
                  yearly_seasonality=True,
                  weekly_seasonality=True, daily_seasonality=True,
                  changepoint_prior_scale=0.5)
prophet.fit(df)

# periods 일을 내다보며 예측합니다.
future_data = prophet.make_future_dataframe(periods=7, freq='d')
forecast_data = prophet.predict(future_data)
forecast_data[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(5)

# #예측값 찍어보기
# fig1 = prophet.plot(forecast_data)

# 예측 넘버 추출
print('매직넘버: ', int(forecast_data.iloc[[forecast_data.shape[0]-1], 1]))

df = full_df[['date', 'Num1']]
df.columns = ['ds', 'y']
prophet = Prophet(seasonality_mode='multiplicative',
                  yearly_seasonality=False,
                  weekly_seasonality=False, daily_seasonality=False,
                  changepoint_prior_scale=0.5)
prophet.fit(df)

# periods 일을 내다보며 예측합니다.
future_data = prophet.make_future_dataframe(periods=7, freq='d')
forecast_data = prophet.predict(future_data)
forecast_data[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(5)

# #예측값 찍어보기
# fig1 = prophet.plot(forecast_data)

# 시작 넘버 추출
print('시작번호:', int(forecast_data.iloc[[forecast_data.shape[0]-1], 1]))

df = full_df[['date', 'Num2']]
df.columns = ['ds', 'y']
prophet = Prophet(seasonality_mode='multiplicative',
                  yearly_seasonality=False,
                  weekly_seasonality=False, daily_seasonality=False,
                  changepoint_prior_scale=0.5)
prophet.fit(df)

# periods 일을 내다보며 예측합니다.
future_data = prophet.make_future_dataframe(periods=7, freq='d')
forecast_data = prophet.predict(future_data)
forecast_data[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(5)

# #예측값 찍어보기
# fig1 = prophet.plot(forecast_data)

# 시작 넘버 추출
print('시작번호:', int(forecast_data.iloc[[forecast_data.shape[0]-1], 1]))

df = full_df[['date', 'Num3']]
df.columns = ['ds', 'y']
prophet = Prophet(seasonality_mode='multiplicative',
                  yearly_seasonality=False,
                  weekly_seasonality=False, daily_seasonality=False,
                  changepoint_prior_scale=0.5)
prophet.fit(df)

# periods 일을 내다보며 예측합니다.
future_data = prophet.make_future_dataframe(periods=7, freq='d')
forecast_data = prophet.predict(future_data)
forecast_data[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(5)

# #예측값 찍어보기
# fig1 = prophet.plot(forecast_data)

# 시작 넘버 추출
print('시작번호:', int(forecast_data.iloc[[forecast_data.shape[0]-1], 1]))

df = full_df[['date', 'Num4']]
df.columns = ['ds', 'y']
prophet = Prophet(seasonality_mode='multiplicative',
                  yearly_seasonality=False,
                  weekly_seasonality=False, daily_seasonality=False,
                  changepoint_prior_scale=0.5)
prophet.fit(df)

# periods 일을 내다보며 예측합니다.
future_data = prophet.make_future_dataframe(periods=7, freq='d')
forecast_data = prophet.predict(future_data)
forecast_data[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(5)

# #예측값 찍어보기
# fig1 = prophet.plot(forecast_data)

# 시작 넘버 추출
print('시작번호:', int(forecast_data.iloc[[forecast_data.shape[0]-1], 1]))

df = full_df[['date', 'Num5']]
df.columns = ['ds', 'y']
prophet = Prophet(seasonality_mode='multiplicative',
                  yearly_seasonality=False,
                  weekly_seasonality=False, daily_seasonality=False,
                  changepoint_prior_scale=0.5)
prophet.fit(df)

# periods 일을 내다보며 예측합니다.
future_data = prophet.make_future_dataframe(periods=7, freq='d')
forecast_data = prophet.predict(future_data)
forecast_data[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(5)

# #예측값 찍어보기
# fig1 = prophet.plot(forecast_data)

# 시작 넘버 추출
print('시작번호:', int(forecast_data.iloc[[forecast_data.shape[0]-1], 1]))

df = full_df[['date', 'Num6']]
df.columns = ['ds', 'y']
prophet = Prophet(seasonality_mode='multiplicative',
                  yearly_seasonality=False,
                  weekly_seasonality=False, daily_seasonality=False,
                  changepoint_prior_scale=0.5)
prophet.fit(df)

# periods 일을 내다보며 예측합니다.
future_data = prophet.make_future_dataframe(periods=7, freq='d')
forecast_data = prophet.predict(future_data)
forecast_data[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(5)

# #예측값 찍어보기
# fig1 = prophet.plot(forecast_data)

# 시작 넘버 추출
print('시작번호:', int(forecast_data.iloc[[forecast_data.shape[0]-1], 1]))
```
